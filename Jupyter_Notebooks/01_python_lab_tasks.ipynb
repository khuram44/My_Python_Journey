```python
num = int(input("Enter a number: "))
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("Number is Prime")
else:
    print("Number is Not Prime")
```

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result = result * i
    return result

num = int(input("Enter a number: "))
print("Factorial is:", factorial(num))
```

```python
students = (
    (1, "khuram", (78, 85, 83)),
    (2, "miksi", (79, 86, 90)),
    (3, "soffi", (71, 84, 77)),
    (4, "ali", (60, 69, 76)),
    (5, "zaki", (58, 65, 81))
)
highest_avg = 0
top_students = []
count_above_75 = 0
for student in students:
    roll, name, marks = student
    avg = sum(marks) / 3
    print(name, "average is :", avg)
    if avg > highest_avg:
        highest_avg = avg
        top_students = [name]
    elif avg == highest_avg:
        top_students.append(name)
    if avg > 75:
        count_above_75 += 1
print("highest average students is :", top_students)
print("students above 75 average is :", count_above_75)
updated_students = (
   (1, "khuram", (78, 85, 83)),
    (2, "miksi", (79, 86, 90)),
    (3, "soffi", (71, 84, 77)),
    (4, "ali", (60, 69, 76)),
    (5, "zaki", (58, 65, 81))
)

print("the updated records is :", updated_students)
```
